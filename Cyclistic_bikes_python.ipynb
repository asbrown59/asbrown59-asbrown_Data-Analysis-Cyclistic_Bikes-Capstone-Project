```python
import pandas as pd

# Assuming df is your DataFrame containing ride data
# Convert 'started_at' column to datetime object
df['started_at'] = pd.to_datetime(df['started_at'])

# Extract day of the week (0=Monday, 1=Tuesday, ..., 6=Sunday)
df['day_of_week'] = df['started_at'].dt.dayofweek

# Map day of the week to actual day names
day_names = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
df['day_of_week'] = df['day_of_week'].map(lambda x: day_names[x])

# Group the DataFrame by member_casual and day_of_week and count occurrences
usage_by_rider_type = df.groupby(['member_casual', 'day_of_week']).size()

# Find the day of the week with the highest usage for members and casual riders
max_usage_by_rider_type = usage_by_rider_type.groupby(level=0).idxmax()

# Display the results
print("Day of the week with the highest usage for members:", max_usage_by_rider_type['member'])
print("Day of the week with the highest usage for casual riders:", max_usage_by_rider_type['casual'])
```

```python
import pandas as pd
from math import radians, sin, cos, sqrt, atan2

# Function to calculate distance between two points using Haversine formula
def calculate_distance(lat1, lon1, lat2, lon2):
    # Convert latitude and longitude from degrees to radians
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])

    # Radius of the Earth in kilometers
    R = 6371.0

    # Haversine formula
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    distance = R * c

    return distance

# Example DataFrame with ride data (replace this with your actual DataFrame)
data = {
    'member_casual': ['member', 'casual', 'member'],
    'start_lat': [40.7128, 34.0522, 40.7128],
    'start_lon': [-74.0060, -118.2437, -74.0060],
    'end_lat': [34.0522, 37.7749, 34.0522],
    'end_lon': [-118.2437, -122.4194, -118.2437]
}
df = pd.DataFrame(data)

# Calculate distance for each ride
df['distance_km'] = df.apply(lambda row: calculate_distance(row['start_lat'], row['start_lon'], row['end_lat'], row['end_lon']), axis=1)

# Aggregate distances for members and casual riders separately
average_distance_by_rider_type = df.groupby('member_casual')['distance_km'].mean()

# Display the average distance traveled by members and casual riders
print("Average distance for members:", average_distance_by_rider_type['member'], "km")
print("Average distance for casual riders:", average_distance_by_rider_type['casual'], "km")
```
